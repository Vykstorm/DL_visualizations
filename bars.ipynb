This notebook shows how to use matplotlib to plot horizontal / vertical bar plots <br/>
Im going to use data i downloaded from Kaggle: https://www.kaggle.com/gregorut/videogamesales

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap


%matplotlib inline
```

```python
data = pd.read_csv('data/vgsales.csv')
data = data[['Name', 'Platform', 'Year', 'Publisher', 'EU_Sales', 'JP_Sales']]
data.rename(columns=str.lower, inplace=True)
data.head()
```

The next plot displays the number of games released on each platform using a vertical bar graphic

```python
plt.figure(figsize=(11, 5))

colors = get_cmap('jet')(np.linspace(0, 1, len(data['platform'].unique())))

data['platform'].map(str.lower).value_counts().sort_values(ascending=False).plot.bar(
    width=0.95, color=colors, align='center')
plt.title('Number of videogames released on each platform')
plt.xlabel('Platform')
plt.ylabel('Videogames')
plt.yticks(np.arange(0, 2200, 200));
```

Number of videogames released on each year (after 2000), this time the plot prints veritical lines

```python
plt.figure(figsize=(11, 5))

years = data[data['year'] > 2000]['year'].astype(np.uint16)
colors = get_cmap('tab20b')(np.linspace(0, 1, len(years.unique())))

years.value_counts().sort_index(ascending=False).plot.barh(
    width=0.9, color=colors, align='center')
plt.title('Number of videogames released on each year')
plt.xlabel('Videogames')
plt.ylabel('Year')
plt.grid(axis='x')
plt.xticks(np.arange(0, 1400, 200));
```
