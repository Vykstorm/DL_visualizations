Here we will plot 3D surfaces

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
import seaborn as sns
from sklearn.datasets import make_classification
from mpl_toolkits.mplot3d import Axes3D
```

First we will define a grid of points over the xy plane. Then assign to each one, a value in the 3th dimension

```python
xx, yy = np.meshgrid(np.linspace(-4, 4, 50), np.linspace(-4, 4, 50))
```

```python
z = np.sqrt((np.divide(1, 1 + xx ** 2) * np.divide(1, 1 + yy ** 2))) * (np.sin(xx) + np.cos(yy))
```

We can first make a scatter plot on the 3d space

```python
fig = plt.figure(figsize=(8, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xx, yy, z, marker='8', s=8)
ax.autoscale_view()
ax.view_init(35, 40)
```

We can instead plot the wireframed surface

```python
fig = plt.figure(figsize=(8, 7))
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(xx, yy, z, rstride=2, cstride=2)
ax.autoscale_view()
ax.view_init(45, 140)
```

Or a solid surface painted

```python
fig = plt.figure(figsize=(8, 7))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(xx, yy, z, rstride=1, cstride=1, cmap='plasma')
ax.autoscale_view()
ax.view_init(45, 150)
```
