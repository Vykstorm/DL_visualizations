This notebook shows how to draw simple animations using matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML

%matplotlib inline
```

```python
x = np.linspace(0, 2*np.pi, 100)
y, y2 = np.sin(x), np.cos(x)
```

We are going to animate the graphic below

```python
plt.figure(figsize=(7, 4))
plt.plot(x, y, linewidth=2, color='blue')
plt.plot(x, y2, linewidth=2, color='red')
```

Create the figure & the plots for each function we want to draw  (when calling plot() we get a sequence where the first item is an instance of class Line2D)

```python
fig, ax = plt.subplots(figsize=(7, 4))
ax.set_xlim((0, 2*np.pi))
ax.set_ylim((-1, 1))
lines, = ax.plot([], [], linewidth=2, color='blue')
lines2, = ax.plot([], [], linewidth=2, color='red')
```

We need to define two functions. init() and animate(). In animate() we call the method set_data() on the Lines2D instances to update the plotting for both functions

```python
def init():
    return (lines, lines2)

def animate(i):
    xx = x + (i / 100) * 2 * np.pi
    y, y2 = np.sin(xx), np.cos(xx)
    lines.set_data(x, y)
    lines2.set_data(x, y2)
    return (lines, lines2)
```

Now create a FuncAnimation instance. We can setup few animations properties like number of frames, speed, ...

```python
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=True)
```

Finally we display the animation

```python
HTML(anim.to_html5_video())
```
