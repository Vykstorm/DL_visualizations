This notebook shows how to create a simple plot to represent a time series

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

The next graph plots 3 functions with different colors, line styles and line widths 

```python
x = np.linspace(1, 10, 50)
y1 = np.log(x)
y2 = np.sin(x)
y3 = 0.5 * np.cos(x)

plt.plot(x, y, color='blue', linewidth=2, linestyle='-')
plt.plot(x, y2, color='red', linewidth=1, linestyle='--', alpha=0.65)
plt.plot(x, y3, color='green', linewidth=2, linestyle=':', alpha=0.65)

plt.legend(['log(x)', 'sin(x)', 'cos(x) / 2'])
plt.xticks(np.linspace(1, 10, 10));
```

Now we do the same graph but indicating line style and colors with the 'fmt' parameter

```python
plt.plot(x, y, 'b-', linewidth=2)
plt.plot(x, y2, 'r--', linewidth=1, alpha=0.65)
plt.plot(x, y3, 'g:', linewidth=2, alpha=0.65)
plt.legend(['log(x)', 'sin(x)', 'cos(x) / 2'])
plt.xticks(np.linspace(1, 10, 10));
```

The graph below displays a grid with vertical & horizontal lines under the plotted function

```python
x = np.linspace(-5, 5, 25)
y = 1 / (1 + np.e ** -x)
plt.plot(x, y)
plt.xticks(np.arange(-6, 6))
plt.yticks(np.linspace(0, 1, 5))
plt.grid(axis='both')
plt.legend(['sigmoid(x)']);
```

The next plot draws horizontal lines indicating the function minimum & maximum values.

```python
x = np.linspace(0, 2*np.pi, 60)
y = np.sin(x) * np.cos(x)
plt.plot(x, y, 'b-')
plt.hlines([y.min(), y.max()], xmin=x.min(), xmax=x.max(), linestyles='--', colors='red', alpha=0.6);
plt.hlines(0, xmin=x.min(), xmax=x.max(), linestyles='-', colors='black', alpha=0.25)
plt.xticks(np.deg2rad(np.arange(0, 360+90, 90)), ['{}º'.format(x) for x in np.arange(0, 360+90, 90)])
plt.grid(axis='x')
```

We can also use the helper functions we have in pandas library in case where we need to plot a pandas series

```python
x = np.linspace(0, 2*np.pi, 50)
y, y2, y3 = np.sin(x), 0.5 * np.sin(x), np.cos(x)
df = pd.DataFrame({'sin(x)':y, 'sin(x) / 2': y2, 'cos(x)': y3}, index=x)
df.plot()

plt.grid(axis='x')
plt.xticks(np.deg2rad(np.arange(0, 360+90, 90)), ['{}º'.format(x) for x in np.arange(0, 360+90, 90)]);
```
